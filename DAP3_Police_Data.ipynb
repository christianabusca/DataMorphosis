# **Data Analytics Project 3 - Police Data**

```python
# IMPORT LIBRARIES HERE
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/DataAnalyticsProject/police_data.csv')
data
```

```python
# 1. REMOVING THE COLUMN THAT ONLY CONTAINS MISSING VALUES
# df.isnull().sum()
data.isnull().sum()
```

```python
# df.drop(column = 'Column_Name', inplace = True)
data.drop(columns = 'country_name', inplace = True) # IF WE ARE GOING TO REMOVE ROWS CHANGE COLUMNS INTO 'index'
# inplace = True permanently removes the data
data
```

```python
# 2. FOR SPEEDING, WERE MEN OR WOMEN STOPPED MORE OFTEN?
# df[df.Column_1 == 'Element/Value'].Column_2,value_counts()
data.head()
data[data.violation == 'Speeding'].driver_gender.value_counts()
# Based on the data shows that Male often has a violation for speeding
```

```python
# 3. Does Gender affect who gets searched during a stop?
data.groupby('driver_gender').search_conducted.sum()
# BASED ON THE DATA, YES
```

```python
data.search_conducted.value_counts()
```

```python
# 4. What is the mean of the stop_duration?
# df['Column_Name'] = df['Column_Name'].map((old:new, old:new))
# df['Column_Name'].mean()
data.head()

data['stop_duration'] = data['stop_duration'].map({'0-15 Min' : 7.5, '16-30 Min' : 24, '30+ Min' : 45})
data.head()
```

```python
data['stop_duration'].mean()
```

```python
# COMPARE THE AGE DISTRIBUTION FOR EACH VIOLATION
# df.groupby('Column_1').Column_2.describe()
data.groupby('violation').driver_age.describe()
```

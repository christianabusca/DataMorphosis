```python
# EXTRACTING THE FILE FROM ZIP FILE
import zipfile
zip_ref = zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall() # THIS WILL EXTRACT THE FILE TO DIR
zip_ref.close() # THIS WILL CLOSE THE FILE
```

```python
# READ THE DATA FROM THE FILE AND HANDLE THE NULL VALUES
import pandas as pd
retail = pd.read_csv('orders.csv', na_values=['Not Available', 'unknown'])
retail['Ship Mode'].unique() #
```

```python
# RENAMING ALL COLUMNS BASED ON PROPER PRACTICE
# WHERE IT SHOULD ALL BE IN LOWERCASE AND NO SPACE CHANGING IT FROM UNDERSCORE
# retail.rename(columns={'Order Id':'order_id','City':'city'})

# FAST WAY
retail.columns = retail.columns.str.lower()
retail.columns = retail.columns.str.replace(' ', '_')
retail.head()
```

```python
# DERIVING NEW COLUMNS DISCOUNT, SALE PRICE, AND PROFIT
retail['discount'] = retail['list_price']*retail['discount_percent']*.01
retail['sale_price'] = retail['list_price']-retail['discount']
retail['profit'] = retail['sale_price'] - retail['cost_price']
retail
```

```python
# CONVERT THE DATE COLUMNS INTO A DATE TYPES
retail.dtypes # THE ORDER_DATE is an Object
retail['order_date'] = pd.to_datetime(retail['order_date'], format='%Y-%m-%d')
retail.head()
```

```python
# DROP THE COST PRICE, LIST PRICE AND DISCOUNT PERCENT COLUMNS
retail.drop(columns=['cost_price', 'list_price', 'discount_percent'], inplace = True)
retail
```

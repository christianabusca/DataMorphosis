```python
# PROBLEM NO. 1: FIND ALL THE UNIQUE 'WIND SPEED' VALUES IN THE DATA
data['Wind Speed_km/h'].unique()
```

```python
# PROBLEM NO.2 : FIND THE NUMBER OF TIMES WHEN THE 'WEATHER' IS EXACTLY CLEAR
# Using Value Counts
data[data['Weather'] == 'Clear'].value_counts()
```

```python
# THROUGH FILTERING
data[data['Weather'] == 'Clear'].filter(['Weather']).count()
```

```python
#THROUGH GROUPBY()
data.groupby('Weather').get_group('Clear').count()
data[data['Weather'] == 'Clear'].groupby('Weather').count()
```

```python
# PROBLEM 3: FIND THE NUMBER OF TIMES WHE THE "WIND SPEED WAS EXACTLY 4KM/H"

data[data['Wind Speed_km/h'] == 4].count( ) # 474
data[data['Wind Speed_km/h'] == 4].value_counts() # 474
data[data['Wind Speed_km/h'] == 4].filter(['Wind Speed_km/h']).count() # 474, MUCH BETTER
data[data['Wind Speed_km/h'] == 4].groupby('Wind Speed_km/h').count()
```

```python
# PROBLEM 4: FIND OUT ALL THE NULL VALUES IN THE DATA
data.isnull().sum() # SHOWS THE NUMBER OF NULL VALUES IN EACH COLUMN
```

```python
# PROBLEM NO. 5: RENAME THE COLUMN NAME "WEATHER" TO "WEATHER CONDITION"

data.rename(columns = {'Weather' : 'Weather Condition'}, inplace = True) # RENAME THE COLUMN 'WEATHER' TO 'WEATHER CONDITION'
data.head()
```

```python
# PROBLEM NO. 6: WHAT IS THE MEAN OF THE VISIBILITY?

data['Visibility_km'].mean() # MEAN OF THE VISIBILITY
```

```python
# PROBLEM NO. 7: WHAT IS THE STANDARD DEVIATION OF THE PRESSURE AREA?

data['Press_kPa'].std() # STANDARD DEVIATION OF THE PRESSURE AREA
```

```python
# PROBLEM NO. 8: FINDING THE VARIANCE OF THE HUMIDITY
data['Rel Hum_%'].var() # VARIANCE OF THE HUMIDITY
```

```python
# PROBLEM NO. 9: FINDING ALL INSTANCES WHERE SNOW IS RECORDED
data.head()
data[data['Weather Condition'].str.contains('Snow')] # SHOWS ALL THE INSTANCES WHERE SNOW IS RECORDED
data[data['Weather Condition'].str.contains('Snow')].filter(['Weather Condition']).count() # SHOWS THE NUMBER OF ALL INSTANCES WHERE SNOW IS RECORDED
```

```python
# PROBLEM NO. 10: FIND ALL INSTANCES WHEN 'WIND SPEED' IS ABOVE 24 AND 'VISIBILITY' IS 25

data[(data['Wind Speed_km/h'] > 24) & (data['Visibility_km'] == 25)].filter(['Wind Speed_km/h', 'Visibility_km']).count()
data[(data['Wind Speed_km/h'] > 24) & (data['Visibility_km'] == 25)]
```
